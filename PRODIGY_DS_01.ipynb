```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("population.csv")
data.head(266)
```

```python
data.describe
```

```python
# Define a set of countries to visualize
selected_countries = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France']

# Extract rows where the country is in our selected list
subset = data[data['Country Name'].isin(selected_countries)]

# Plot a bar graph for population in the year 2022
plt.figure(figsize=(8,5))
plt.bar(subset['Country Name'], subset['2022'], color='steelblue')
plt.title('2022 Population of Selected Nations')
plt.xlabel('Country')
plt.ylabel('Population (2022)')
plt.xticks(rotation=45, ha='right')  # Rotate labels to avoid overlap
plt.tight_layout()
plt.show()
```

```python
# Example: Plot population for 2020, 2021, 2022 if available
years = ['2020', '2021', '2022']

plt.figure(figsize=(8,6))
for _, row in subset.iterrows():
    plt.plot(years, row[years], marker='o', label=row['Country Name'])

plt.title('Population Trend (2020-2022)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(subset['2021'], subset['2022'], color='darkorange')
for i, row in subset.iterrows():
    plt.text(row['2021'], row['2022'], row['Country Name'], fontsize=8)

plt.title('Population Change from 2021 to 2022')
plt.xlabel('Population in 2021')
plt.ylabel('Population in 2022')
plt.grid(True)
plt.tight_layout()

plt.show()
```
